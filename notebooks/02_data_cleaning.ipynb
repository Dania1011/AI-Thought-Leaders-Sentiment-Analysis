```python
import re

def clean_text(text):
    if pd.isna(text):
        return ""
    text = str(text)
    text = re.sub(r"http\S+|www.\S+", "", text)        # remove URLs
    text = re.sub(r"@[A-Za-z0-9_]+", "", text)         # remove mentions
    text = re.sub(r"[^A-Za-z0-9\s]", "", text)         # remove emojis/symbols
    text = text.lower().strip()                        # normalize
    return text

# Create clean_text column
df["clean_text"] = df["text"].apply(clean_text)

# Remove empty rows
df = df[df["clean_text"] != ""]

print("After cleaning:", df.shape)
```

```python
df.drop_duplicates(subset=["tweet_id", "created_at"], keep="first", inplace=True)
```

```python
df.drop_duplicates(subset=["text", "created_at"], keep="first", inplace=True)
```

```python
df = df[
    (df["is_retweet"] == False) &
    (df["is_reply"] == False) &
    (df["is_quote"] == False)
]
```

```python
df = df[df["lang"] == "en"]
```

```python
df = df[df["text"].notna()]
df = df[df["text"].str.strip() != ""]
```
